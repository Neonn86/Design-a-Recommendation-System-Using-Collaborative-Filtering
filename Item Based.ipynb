```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Combined.csv")
df = df.melt(id_vars=['User First Name', 'User Last Name'])
df.rename(columns={'variable':'Movie Name', 'value':'Rating'}, inplace=True)
df.drop("User Last Name",axis=1,inplace=True)

df['Movie Name'] = df['Movie Name'].str.strip()

movies_drop = ["The Social Network", "Mad Max: Fury Road", "Inception"]
users_drop = ["Kavish", "Jeet", "Neon", "Keshore", "Asha"]

for user in users_drop:
    for movie in movies_drop:
        filter = df.loc[(df['User First Name'] == user) & (df['Movie Name'] == movie), "Rating"]
        df.replace(filter,np.nan,inplace=True)   
df=df.pivot(index='User First Name',columns="Movie Name")
df
# df["UserID"]=range(280)
# df=df.drop('User Last Name',axis=1)
# df=df.set_index('User First Name')
# df=df.T
```

```python
df.columns = [i[1] for i in df.columns]
```

```python
df.columns
```

## Calculation

```python
df1=df.copy()
```

```python
mean_df1=df1.mean(axis=0)
mean_df1
```

```python
std_df1=df1.std()
std_df1
```

```python
df1=(df1-mean_df1)/std_df1
df1.head()
```

```python
df1.fillna(0, inplace = True)
df1.head()
```

```python
df1.shape
```

## Cosine Similarity

```python
df_cosine=pd.DataFrame()

Our_Movies=["The Social Network", "Mad Max: Fury Road", "Inception"]

Movies=list(df1.columns)

from scipy import spatial

for our_movie in Our_Movies:
    for movie in Movies:
        result = 1 - spatial.distance.cosine(df1[our_movie], df1[movie])
        df_cosine=df_cosine.append({'Movie_1': our_movie,"Movie_2":movie,"Cosine_Sim":result}, ignore_index=True)
df_cosine
    
```

```python
df_cosine.shape
```

```python
df_cosine.drop(df_cosine[df_cosine['Movie_1'] == df_cosine['Movie_2']].index, inplace = True)
df_cosine["Abs_cosine"]=df_cosine["Cosine_Sim"].abs()
df_cosine.shape
```

```python
df_cosine
```

```python
dict_cosine_sum={}
for our_movie in Our_Movies:
    total_ten=df_cosine[df_cosine["Movie_1"]== our_movie].sort_values(by=['Cosine_Sim'],ascending=False).iloc[:10,]
    #print(total_ten)
    first_ten_sum=total_ten["Abs_cosine"].sum()
    dict_cosine_sum[our_movie]=first_ten_sum

dict_cosine_sum
```

```python
dict_id_rank={}
for our_movie in Our_Movies:
    dict_id_rank[our_movie]=list(df_cosine[df_cosine["Movie_1"]== our_movie].sort_values(by=['Cosine_Sim'],ascending=False)["Movie_2"][:10].values)
dict_id_rank
```

```python
mean_df=pd.DataFrame(mean_df1)


std_df=pd.DataFrame(std_df1)
```

```python
df_predict=pd.DataFrame()
Teams=["Kavish", "Jeet", "Neon", "Keshore", "Asha"]
rating_sum=0
for team in Teams:
    #print("###"+f"{movie}")
    for our_movie,group in dict_id_rank.items():
        rating_sum=0
        #print("###"+f"{member}")
        for movie in group:
            rating=df1.loc[team,movie]
            cosine_s=df_cosine.loc[(df_cosine["Movie_1"]==our_movie) & (df_cosine["Movie_2"]==movie)]["Cosine_Sim"].values[0]
            #print(f"{people},{rating},{cosine_s}")
            rating_sum+= rating * cosine_s
            #print(rating_sum)
        #print(rating_sum)
        #print(f"{movie},{member},{rating_sum}")
        df_predict=df_predict.append({'Name':team,'Movie': our_movie,"Rating":(rating_sum / dict_cosine_sum[our_movie] * std_df.loc[our_movie,0]) + mean_df.loc[our_movie,0]}, ignore_index=True)
df_predict
```
