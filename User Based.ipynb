```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("Combined.csv")
df = df.melt(id_vars=['User First Name', 'User Last Name'])
df.rename(columns={'variable':'Movie Name', 'value':'Rating'}, inplace=True)
df.drop("User Last Name",axis=1,inplace=True)

df['Movie Name'] = df['Movie Name'].str.strip()

movies_drop = ["The Social Network", "Mad Max: Fury Road", "Inception"]
users_drop = ["Kavish", "Jeet", "Neon", "Keshore", "Asha"]

for user in users_drop:
    for movie in movies_drop:
        filter = df.loc[(df['User First Name'] == user) & (df['Movie Name'] == movie), "Rating"]
        df.replace(filter,np.nan,inplace=True)   
df=df.pivot(index='Movie Name',columns='User First Name')
# df["UserID"]=range(280)
# df=df.drop('User Last Name',axis=1)
# df=df.set_index('User First Name')
# df=df.T
```

```python
df.columns = [i[1] for i in df.columns]
```

```python
df.columns
```

## Calculation

```python
df1=df.copy()
```

```python
mean_df1=df1.mean(axis=0)
mean_df1
```

```python
std_df1=df1.std()
std_df1
```

```python
df1=(df1-mean_df1)/std_df1
df1.head()
```

```python
df1.fillna(0, inplace = True)
df1.head()
```

```python
df1.shape
```

## Cosine Similarity

```python
df_cosine=pd.DataFrame()

Teams=["Kavish", "Jeet", "Neon", "Keshore", "Asha"]

Names=list(df1.columns)

from scipy import spatial

for team in Teams:
    for name in Names:
        result = 1 - spatial.distance.cosine(df1[team], df1[name])
        df_cosine=df_cosine.append({'User_1': team,"User_2":name,"Cosine_Sim":result}, ignore_index=True)
df_cosine
    
```

```python
df_cosine.shape
```

```python
df_cosine.drop(df_cosine[df_cosine['User_1'] == df_cosine['User_2']].index, inplace = True)
df_cosine["Abs_cosine"]=df_cosine["Cosine_Sim"].abs()
df_cosine.shape
```

```python
df_cosine
```

```python
dict_cosine_sum={}
for i in Teams:
    total_ten=df_cosine[df_cosine["User_1"]== i].sort_values(by=['Cosine_Sim'],ascending=False).iloc[:10,]
    #print(total_ten)
    first_ten_sum=total_ten["Abs_cosine"].sum()
    dict_cosine_sum[i]=first_ten_sum

dict_cosine_sum
```

```python
dict_id_rank={}
for team in Teams:
    dict_id_rank[team]=list(df_cosine[df_cosine["User_1"]== team].sort_values(by=['Cosine_Sim'],ascending=False)["User_2"][:10].values)
dict_id_rank
```

```python
# total_ten=df_cosine[df_cosine["User_1"]== "Neon"].sort_values(by=['Cosine_Sim'],ascending=False).iloc[:10,]
a=[]
for people in dict_id_rank["Neon"]:
    a.append(df1.loc["Inception",people])
total_ten["Rating"]=a
total_ten
```

```python
dict_id_rank={}
for i in Teams:
    dict_id_rank[i]=list(df_cosine[df_cosine["User_1"]== i].sort_values(by=['Cosine_Sim'],ascending=False)["User_2"][:10].values)
dict_id_rank
```

```python
mean_df=pd.DataFrame(mean_df1)


std_df=pd.DataFrame(std_df1)
```

```python
df_predict=pd.DataFrame()
Movies=["The Social Network", "Mad Max: Fury Road", "Inception"]
rating_sum=0
for movie in Movies:
    #print("###"+f"{movie}")
    for member,group in dict_id_rank.items():
        rating_sum=0
        #print("###"+f"{member}")
        for people in group:
            rating=df1.loc[movie,people]
            cosine_s=df_cosine.loc[(df_cosine["User_1"]==member) & (df_cosine["User_2"]==people)]["Cosine_Sim"].values[0]
            #print(f"{people},{rating},{cosine_s}")
            rating_sum+= rating * cosine_s
            #print(rating_sum)
        #print(rating_sum)
        #print(f"{movie},{member},{rating_sum}")
        df_predict=df_predict.append({'Movies':movie,'Name': member,"Rating":(rating_sum / dict_cosine_sum[member] * std_df.loc[member,0]) + mean_df.loc[member,0]}, ignore_index=True)
df_predict
```

```python
df_predict_p=df_predict.pivot(index='Movies',columns='Name')
df_predict_p.to_csv(r'C:\Users\win10\Desktop\User-based.csv',index=True)

```
